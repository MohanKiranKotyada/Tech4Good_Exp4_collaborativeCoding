Now we try and assign codes to each student for peer review

```python
import random
import math

# Number of students
N = 9

# Calculate X = sqrt(N)
X = math.isqrt(N)  # Round to the nearest integer

# Initialize the review assignments dictionary
reviews = {f'user_{i}': [] for i in range(1, N+1)}
```

Advantage of using the "square root" meathod is that we get to assign all the students with same amount of codes, and each code is reviwed almost by the same number of students thus ensuring there is no bias while aggregating the results.

However, in a scenerio where there are more than 50 students, where each student gets 7 codes to review, it becomes tough for students and a better solution needs to be though of.

```python
# Function to distribute reviews randomly but fairly
def assign_reviews():
    # Create a list of all user IDs
    all_users = [f'user_{i}' for i in range(1, N+1)]

    # Iterate over each student to assign codes for review
    for user in all_users:
        # Make a copy of all users and remove the current user (to avoid self-review)
        available_users = [u for u in all_users if u != user]

        # Randomly choose X users for review
        assigned_users = random.sample(available_users, X)

        # Assign the selected codes to the current user
        reviews[user].extend(assigned_users)

    # Ensure balanced review
    # Iterate and check if any code is not reviewed by X students, redistribute if necessary
    for user, assigned_users in reviews.items():
        for assigned_user in assigned_users:
            if len([u for u, r in reviews.items() if assigned_user in r]) < X:
                continue
            # If a code has been reviewed enough times, redistribute
            replacement = random.choice([u for u in all_users if u != user and u not in assigned_users])
            assigned_users.remove(assigned_user)
            assigned_users.append(replacement)

# Assign reviews
assign_reviews()
```

```python
#Showing which student should review which code
for user, review_list in reviews.items():
    print(f"{user} will review the codes of: {', '.join(review_list)}")
```

Menu where each user should review others code (3 in this case). Please do note that I have randomly reviwed the codes, i.e., I did not actually read the code and mark. I have just marked the code with random values.

```python
reviews_data = {f'user_{i}': {} for i in range(1, N+1)}

# Function to perform the review process for each reviewer
def review_code(reviewer, assigned_users):
    print(f"\nReviewer {reviewer}, please come forward and review the assigned codes.\n")
    for user in assigned_users:
        print(f"Reviewing code of {user}")

        # Input the scores based on rubrics
        correctness = float(input("Correctness (out of 5): "))
        efficiency = float(input("Efficiency (out of 2): "))
        readability = float(input("Readability (out of 1.5): "))
        robustness = float(input("Robustness (out of 1): "))
        originality = float(input("Originality (out of 0.5): "))

        # Calculate total score
        total_score = correctness + efficiency + readability + robustness + originality

        # Input feedback
        feedback = input("Please provide your feedback: ")

        # Store the results in the reviews_data structure
        reviews_data[reviewer][user] = {
            'correctness': correctness,
            'efficiency': efficiency,
            'readability': readability,
            'robustness': robustness,
            'originality': originality,
            'total_score': total_score,
            'feedback': feedback
        }
        print("\n")

# Execute the review process for each reviewer
for i in range(1, N+1):
    reviewer = f'user_{i}'  # Corrected key
    assigned_users = reviews[reviewer]  # Get the assigned users for the current reviewer
    review_code(reviewer, assigned_users)
```

Each user can look at their feedback

```python
# Function to display scores and comments for a user's code
def view_feedback(user):
    print(f"\nFeedback for User {user}'s code:\n")

    # Check if the user has received any reviews
    if user not in reviews_data or not reviews_data[user]:
        print(f"No reviews found for User {user}'s code.\n")
        return

    # Display all reviews
    for reviewer, feedback in reviews_data.items():
        if user in feedback:
            print(f"Reviewed by {reviewer}:")
            print(f"  Correctness: {feedback[user]['correctness']} / 5")
            print(f"  Efficiency: {feedback[user]['efficiency']} / 2")
            print(f"  Readability: {feedback[user]['readability']} / 1.5")
            print(f"  Robustness: {feedback[user]['robustness']} / 1")
            print(f"  Originality: {feedback[user]['originality']} / 0.5")
            print(f"  Total Score: {feedback[user]['total_score']} / 10")
            print(f"  Feedback: {feedback[user]['feedback']}\n")

# Menu to select user and view feedback
def feedback_menu():
    while True:
        print("\n--- Feedback Menu ---")
        print("Enter the user number to view their feedback, or type 'exit' to quit.")
        user_input = input("Enter user number (1-9) or 'exit': ").strip()

        if user_input.lower() == 'exit':
            print("Exiting the feedback menu.")
            break

        if user_input.isdigit():
            user_num = int(user_input)
            if 1 <= user_num <= N:
                view_feedback(f'user_{user_num}')
            else:
                print("Invalid user number. Please enter a number between 1 and 10.")
        else:
            print("Invalid input. Please enter a valid user number or 'exit'.")

# Initialize the menu
feedback_menu()
```

Now we shall aggeragte and find the top 5 codes and their feedbacks

```python
# Function to calculate and return the top 5 codes based on average scores
def get_top_5_codes():
    avg_scores = {}

    # Calculate the average score for each user
    for user in reviews_data:
        total_score = 0
        num_reviews = 0

        for reviewer, feedback in reviews_data.items():
            if user in feedback:
                total_score += feedback[user]['total_score']
                num_reviews += 1

        if num_reviews > 0:
            avg_scores[user] = total_score / num_reviews

    # Sort the users based on their average score in descending order
    sorted_users = sorted(avg_scores.items(), key=lambda x: x[1], reverse=True)

    # Extract the top 5 users
    top_5_users = sorted_users[:5]

    # Display the top 5 users with their scores and feedback
    for rank, (user, avg_score) in enumerate(top_5_users, start=1):
        print(f"\nRank {rank}: {user} with an average score of {avg_score:.2f}/10")
        print("Feedback:")

        for reviewer, feedback in reviews_data.items():
            if user in feedback:
                print(f"  Reviewed by {reviewer}:")
                print(f"    Correctness: {feedback[user]['correctness']} / 5")
                print(f"    Efficiency: {feedback[user]['efficiency']} / 2")
                print(f"    Readability: {feedback[user]['readability']} / 1.5")
                print(f"    Robustness: {feedback[user]['robustness']} / 1")
                print(f"    Originality: {feedback[user]['originality']} / 0.5")
                print(f"    Total Score: {feedback[user]['total_score']} / 10")
                print(f"    Feedback: {feedback[user]['feedback']}\n")

# Calculate and display the top 5 codes
get_top_5_codes()
```

Now we can add a module where the user can compare te scores of his code vs. the top ranked code. Also, there will be personalised message after each rubric telling which code is better in terms of that particular rubric.

In this way, the students can get validated and encouraged when there code is better in atleast one of the rubrics mentioned.

```python
# Function to get the top-ranked code's user and their average scores
def get_top_ranked_code():
    avg_scores = {}

    # Calculate the average score for each user
    for user in reviews_data:
        total_score = 0
        num_reviews = 0

        for reviewer, feedback in reviews_data.items():
            if user in feedback:
                total_score += feedback[user]['total_score']
                num_reviews += 1

        if num_reviews > 0:
            avg_scores[user] = total_score / num_reviews

    # Get the user with the highest average score
    top_user = max(avg_scores, key=avg_scores.get)
    return top_user, avg_scores[top_user]

# Function to compare a user's score with the top-ranked code
def compare_with_top_ranked(user):
    top_user, top_avg_score = get_top_ranked_code()

    print(f"\nComparison of User {user} with Top-Ranked User {top_user}:\n")

    # Check if the user has received any reviews
    if user not in reviews_data or not reviews_data[user]:
        print(f"No reviews found for User {user}'s code.\n")
        return

    # Initialize variables to store user's total scores
    user_total = 0
    num_reviews = 0

    # Calculate user's average score and display differences
    for reviewer, feedback in reviews_data.items():
        if user in feedback:
            user_total += feedback[user]['total_score']
            num_reviews += 1

    user_avg_score = user_total / num_reviews if num_reviews > 0 else 0

    print(f"Your Average Score: {user_avg_score:.2f}/10")
    print(f"Top-Ranked Average Score: {top_avg_score:.2f}/10")
    print(f"Difference: {top_avg_score - user_avg_score:.2f}\n")

    # Compare each rubric category
    for rubric in ['correctness', 'efficiency', 'readability', 'robustness', 'originality']:
        user_rubric_score = sum(feedback[user][rubric] for reviewer, feedback in reviews_data.items() if user in feedback) / num_reviews
        top_rubric_score = sum(feedback[top_user][rubric] for reviewer, feedback in reviews_data.items() if top_user in feedback) / len(reviews_data[top_user])
        difference = top_rubric_score - user_rubric_score

        print(f"{rubric.capitalize()}:")
        print(f"  Your Score: {user_rubric_score:.2f}")
        print(f"  Top-Ranked Score: {top_rubric_score:.2f}")
        print(f"  Difference: {difference:.2f}")

        # Add personalized messages based on the difference
        if difference > 0:
            print(f"  The top code is better than yours in terms of {rubric}.")
        elif difference < 0:
            print(f"  Your code is better than the top code in terms of {rubric}.")
        else:
            print(f"  Your code is equally good as the top code in terms of {rubric}.\n")

def compare_scores_menu():
    while True:
        print("\n--- Compare Scores Menu ---")
        print("Enter your user number to compare your score with the top-ranked code, or type 'exit' to quit.")
        user_input = input("Enter user number (1-10) or 'exit': ").strip()

        if user_input.lower() == 'exit':
            print("Exiting the comparison menu.")
            break

        if user_input.isdigit():
            user_num = int(user_input)
            if 1 <= user_num <= N:
                compare_with_top_ranked(f'user_{user_num}')
            else:
                print("Invalid user number. Please enter a number between 1 and 10.")
        else:
            print("Invalid input. Please enter a valid user number or 'exit'.")

# Initialize the comparison menu
compare_scores_menu()

```

Now, instead of just comparision with the top code, the students should also be able to compare their code, with all the codes that are better than theirs in a particular rubric.

```python
# Function to compare a specific rubric between a user's code and others
def compare_rubric(user, rubric):
    user_rubric_score = sum(feedback[user][rubric] for reviewer, feedback in reviews_data.items() if user in feedback) / len(reviews_data[user])

    print(f"\nPrograms Better Than User {user}'s Code in Terms of {rubric.capitalize()}:\n")
    found_better = False

    for other_user in reviews_data:
        if other_user == user:
            continue

        other_rubric_score = sum(feedback[other_user][rubric] for reviewer, feedback in reviews_data.items() if other_user in feedback) / len(reviews_data[other_user])

        if other_rubric_score > user_rubric_score:
            found_better = True
            print(f"User {other_user}: {other_rubric_score:.2f} (Your Score: {user_rubric_score:.2f})")

    if not found_better:
        print(f"No programs scored better than yours in terms of {rubric.capitalize()}.")

# Function to input user number and select rubric to compare
def rubric_comparison_menu():
    rubric_mapping = {
        1: 'correctness',
        2: 'efficiency',
        3: 'readability',
        4: 'robustness',
        5: 'originality'
    }

    while True:
        print("\n--- Rubric Comparison Menu ---")
        print("Enter your user number and select a rubric to compare.")
        print("Rubrics: 1. Correctness | 2. Efficiency | 3. Readability | 4. Robustness | 5. Originality")
        print("Type 'exit' to quit.")

        user_input = input("Enter user number (1-9) or 'exit': ").strip()

        if user_input.lower() == 'exit':
            print("Exiting the rubric comparison menu.")
            break

        if user_input.isdigit():
            user_num = int(user_input)
            if 1 <= user_num <= N:
                rubric_input = input("Select a rubric by entering its number (1-5): ").strip()

                if rubric_input.isdigit():
                    rubric_num = int(rubric_input)
                    if rubric_num in rubric_mapping:
                        rubric = rubric_mapping[rubric_num]
                        compare_rubric(f'user_{user_num}', rubric)
                    else:
                        print("Invalid rubric number. Please enter a number between 1 and 5.")
                else:
                    print("Invalid input. Please enter a valid rubric number.")
            else:
                print("Invalid user number. Please enter a number between 1 and 10.")
        else:
            print("Invalid input. Please enter a valid user number or 'exit'.")

# Initialize the rubric comparison menu
rubric_comparison_menu()
```

**The users should be able to download and see other uses code, and then compare manually as well**

```python
def download_user_code(user_number):
    # Construct the user key
    user_key = f'user_{user_number}'

    # Check if the user exists in the index
    if user_key in code_index:
        file_path = code_index[user_key][0]

        # Check if the file exists
        if os.path.isfile(file_path):
            # Download the file
            files.download(file_path)
            print(f"Downloading file for {user_key}...")
        else:
            print("File does not exist.")
    else:
        print("User not found in index.")

# Loop to continuously prompt for user input
while True:
    user_input = input("Enter the user number (e.g., 1 for user_1) or type 'stop' to exit: ")

    if user_input.lower() == 'stop':
        print("Stopping the program.")
        break

    try:
        user_number = int(user_input)
        download_user_code(user_number)
    except ValueError:
        print("Invalid input. Please enter a valid user number or 'stop' to exit.")
```
